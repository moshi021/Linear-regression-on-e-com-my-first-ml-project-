```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math, copy
```

```python
data=pd.read_csv('Ecommerce')

X_train = data[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y_train= data['Yearly Amount Spent']
```

```python
from sklearn.linear_model import LinearRegression


linear_model = LinearRegression()
linear_model.fit(X_train, y_train) 
```

```python
b = linear_model.intercept_
w = linear_model.coef_

print(w,b)
```

```python
y_pred = linear_model.predict(X_train)
y_pred
```

```python
plt.scatter(y_train,y_pred)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_train, y_pred)
print(mse)
```

```python
plt.scatter(X_train['Length of Membership'], y_train)
plt.scatter(X_train['Length of Membership'], y_pred, c='r')
```
